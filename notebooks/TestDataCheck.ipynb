```python
import numpy as np
import pylab
```

```python
def plot_figure(data_list, color, label, fig_size=(10, 10)):
    pylab.figure(1)
    plot_value = int('111')
    pylab.figure(figsize=fig_size)
    pylab.subplot(plot_value)
    pylab.xlabel("Time")
    pylab.ylabel("wave")
    for index, data in enumerate(data_list):
        pylab.plot(data, color[index], label=label[index])
    pylab.legend(loc='upper right')
    pylab.show()
    pylab.close()
```

# train X 

```python
Xdata = np.load('../src/tests/test_data/test_X.npy')
data_list = [Xdata]
color_list = ['b']
label_list = ['test_data_train_x']

plot_figure(data_list, color_list, label_list)
```
